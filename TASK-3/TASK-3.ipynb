### EDA and Visualization

```python
#correlation
df.corr()
```

```python
#covariance
df.cov()
```

```python
#data compaarison : heatmap
plt.figure(figsize=(8,6))
sns.heatmap(df.corr(),annot=True,cmap='rocket')
plt.title('Correlation Heatmap')
plt.show()
```

```python
#data comparison (category wise) : pairplot
plt.style.use('ggplot')
plt.figure(dpi=100)
sns.pairplot(df,diag_kind = 'kde')
plt.show()
```

```python
# bar plot : profit on different aspects
plt.style.use('classic')
fig,axes = plt.subplots(2,2,dpi=100,figsize=(13,9))
plt.suptitle('Total Profits',size=15)
sns.barplot(x='Region',y='Profit',data=df,ax=axes[0,0],estimator=np.sum,palette='tab10')
axes[0,0].set_title('Profit vs Region',size=10)
sns.barplot(x='Segment',y='Profit',data=df,ax=axes[0,1],estimator=np.sum,palette='tab10')
axes[0,1].set_title('Segment vs Region',size=10)
sns.barplot(x='Category',y='Profit',data=df,ax=axes[1,0],estimator=np.sum,palette='tab10')
axes[1,0].set_title('Category vs Region',size=10)
sns.barplot(x='Ship Mode',y='Profit',data=df,ax=axes[1,1],estimator=np.sum,palette='tab10')
axes[1,1].set_title('Ship Mode vs Region',size=10)
plt.tight_layout()
plt.show()
```

**OBSERVATION** --> Maximum profits occur from West Region (Region-wise), from Consumer Sector (Segment wise), from Technology (Category-wise) and max profits are earned when shipping mode is Standard Class.

```python
#density plot : profit and sales
plt.style.use('ggplot')
plt.figure(figsize=(7,4))
sns.kdeplot(x='Sales',data=df,label='Sales',shade=True,color='blue')
sns.kdeplot(x='Profit',data=df,label='Profit',shade=True,color='red')
plt.xlim([-500,1000])
plt.legend()
plt.show()
```

**Increasing the orders would result in more profits**

```python
# sub-category wise profits
plt.figure(dpi=100,figsize=(10,6))
sns.barplot(x='Sub-Category',y='Profit',data=df,estimator=np.sum)
plt.title('Sub-Category Profits')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

**OBSERVATION** --> Copiers are providing the maximum profit while Tables are providing losses.

```python
#grouping subcategory under technology
tech = df[df['Category']=='Technology'].groupby(by='Sub-Category').sum()

#plotting pie chart
plt.figure()
plt.pie(x=tech['Sales'],labels=tech.index,shadow=True,colors=['red','blue','brown','yellow'],explode=[0,0.1,0,0],
       autopct='%1.1f%%')
plt.legend(bbox_to_anchor=(1.5,1))
plt.show()
```

**OBSERVATION** --> Copiers have the least sale, yet they earn maximum profits. Increasing the sale of the copiers would result in more profits.

```python
#grouping subcategory under furniture
furniture = df[df['Category']=='Furniture'].groupby(by='Sub-Category').sum()

#plotting pie chart
plt.figure()
plt.pie(x=furniture['Sales'],labels=furniture.index,shadow=True,colors=['red','blue','brown','yellow'],explode=[0,0,0.1,0],
       autopct='%1.1f%%')
plt.legend(bbox_to_anchor=(1.5,1))
plt.show()
```

**OBSERVATION** --> Furnishings have least sale, increasing the sale quantity would earn more profits. Tables selling price should be increased to provide profits.

```python
#grouping subcategory under office suplies
ofs = df[df['Category']=='Office Supplies'].groupby(by='Sub-Category').sum()

#plotting pie chart
plt.figure()
plt.pie(x=ofs['Sales'],labels=ofs.index,shadow=True,explode=[0,0,0,0,0.5,0,0,0,0],autopct='%1.1f%%')
plt.legend(bbox_to_anchor=(1.5,1))
plt.show()
```

**OBSERVATIONS** --> Fasteners are least liked office product and has the minimum sales. The price of supplies should be increase for greater profits.

```python
#discount vs qunatity
plt.figure()
sns.kdeplot(x='Discount',data=df,shade=True,label='Discount',color='red')
sns.kdeplot(x='Quantity',data=df,shade=True,label='Quanity',color='Blue')
plt.title('Discount vs Quantity')
plt.xlim([-4,8])
plt.ylim([0,2])
plt.legend()
```

```python
#STATE WISE ANALYSIS
#state-wise discounts
plt.figure(dpi=100,figsize=(10,8))
sns.barplot(x='State',y='Discount',data=df,estimator=np.sum,palette='tab10')
plt.title('State-wise Discounts')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
#state-wise sales
plt.figure(dpi=100,figsize=(10,8))
sns.barplot(x='State',y='Sales',data=df,estimator=np.sum,palette='tab10')
plt.title('State-wise Sales')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
#state-wise profits
plt.figure(dpi=100,figsize=(10,8))
sns.barplot(x='State',y='Profit',data=df,estimator=np.sum,palette='tab10')
plt.title('State-wise Profits')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

**OBSERVATION** --> discount should be given according to the sale demand to make profits.
- TEXAS offers maximum discount but the sales are less and thus ends up in losses. 
- New York, California has the maximum sales and offers a lil discound making large profits.
- States such as West Virginia offers no discount and ends up in losses.

```python
#CITY-WISE ANALYSIS
city = df.groupby(by='City').sum().sort_values('Profit')

#top 5 cities with losses
city.head()
```

```python
#top 5 cities with profits
city.tail()
```

```python
#profit analysis by city
#loss
main_loss = city.reset_index().head(50)
plt.figure(dpi=100,figsize=(10,7))
sns.barplot(x='City',y='Profit',data=main_loss)
plt.xticks(rotation=90)
plt.title('Top 50 Cities with Losses')
plt.tight_layout()
plt.show()
#profit
main_prof = city.reset_index().tail(50)
plt.figure(dpi=100,figsize=(10,7))
sns.barplot(x='City',y='Profit',data=main_prof)
plt.xticks(rotation=90)
plt.title('Top 50 Cities with Profits')
plt.tight_layout()
plt.show()
```

**OBSERVATION** --> Philedephia suffers maximum losses and New York city earns maximum profits.
